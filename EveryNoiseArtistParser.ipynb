```python
# Slicer Function for style to only include font-size and any other purposes
```

```python
def slicer(my_str,sub):
    index=my_str.find(sub)
    if index !=-1 :
        return my_str[index:] 
    else :
        return "INCORRECT"
```

```python
def artists(artistAndTextSizelist, my_url):
    
    # HTML Parser
    page = requests.get(my_url)
    tree = html.fromstring(page.content)
    
    # For Troubleshooting
    # artistcount1 = 0
    
    # Add in a variable that grabs the genre
    genre = my_url[my_url.find("-")+1:-5]
    
    # Grab All text from webpage
    list1 = tree.xpath('//div[@id]/text()')

    # Delete every other element in list (Makes more sense when reading the html and outputs)
    list2 = list1[0::2]

    # For Troubleshooting
    # artistcount1 = len(thisOne2)
    # All artists are good 
    # print(thisOne2[:447])
    # All genres are good
    # print(thisOne2[-63:])

    # Discern the number of artsits on the webpage.
    # Current process is to find "\n", which happens to be the element always prior to the last artist in the html
    endOfArtists = list2.index("\n")

    # List only artists 
    listArtists = list2[:endOfArtists]
    # print(listArtists)
    listArtists.insert(0,genre)
    artistAndTextSizelist.append(listArtists)
    
    # Call textsize 
    textsize(artistAndTextSizelist, genre, endOfArtists, my_url)
```

```python
def textsize(artistAndTextSizelist, genre, endOfArtists, my_url):
    
    
    # HTML Parser
    page = requests.get(my_url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # For Troubleshooting
    # textsizecount1 = 0

    # Create a list and grab every font size
    list1 = []
    for i in soup.select("div.canvas > div"):
        # textsizecount1=textsizecount1+1
        list1.append(slicer(i.get("style"),"font-size"))

    #Create a list to get all artist Counts    
    
    #For Troubleshooting
    # All artists are good
    # print(list1[:447])
    # All genres are good
    # print(list1[-63:])

    # Limit list to only length of artists
    listTextSizeWithFontAndPercent = list1[:endOfArtists]

    # Remove extra characters from list, so only font size in a number format exists
    # remove "font-size: " from list
    listTextSizeWithPercent = [i[11:] for i in listTextSizeWithFontAndPercent]
    # remove % from list
    listTextSize = [i[:-1] for i in listTextSizeWithPercent]
    # Insert genre
    listTextSize.insert(0,"Text Size for " + genre)
    # print(listTextSize)
    artistAndTextSizelist.append(listTextSize)
#     artistCount.append(str(endofArtists))
    
```
